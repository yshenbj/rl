```python
import struct
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open("./phase20240403", "rb") as file:
    binary_data = file.read()

num_floats = len(binary_data) // 4
float_list = struct.unpack(f"{num_floats}f", binary_data)
data = np.array(float_list).reshape((-1, 500))
data_t = np.array(float_list).reshape((500, -1))
```

```python
i = 4
print("Maximum: ", np.max(data_t[i]))
print("Minimum: ", np.min(data_t[i]))
print("Mean: ", np.mean(data_t[i]))
```

```python
corr = np.corrcoef(data_t[:100, :])
```

```python
axis_corr = sns.heatmap(corr)
```
